```python
with open('input_day10.txt', 'r') as f:
    input_str = f.read()
```

```python
adapters = input_str.splitlines()
adapters = (int(adapter) for adapter in adapters if adapter)
adapters = sorted(adapters)
#print(adapters)
```

```python
# add charging outlet and device
adapters.insert(0, 0)
adapters.append(adapters[-1] + 3)
#print(adapters)
```

```python
import numpy as np

joltage_differences = np.diff(adapters)
#print(joltage_differences)
```

```python
num_1j_differences = np.sum(joltage_differences == 1)
num_3j_differences = np.sum(joltage_differences == 3)
print(num_1j_differences, num_3j_differences, num_1j_differences * num_3j_differences)
```

```python
combination_borders = np.where(joltage_differences == 3)[0]
combination_groups = [joltage_differences[:combination_borders[0]]]
combination_groups += [joltage_differences[combination_borders[i] + 1:combination_borders[i + 1]] for i in range(len(combination_borders) - 1)]
combination_groups = list(filter(len, combination_groups))
#print(combination_groups)
```

```python
def merge_indices(combination_group, indices):
    merged_combination = list(combination_group)
    for i in reversed(sorted(indices)):
        merged_combination[i - 1] = merged_combination[i - 1] + merged_combination[i]
        del merged_combination[i]
    return merged_combination
```

```python
import itertools

num_valid_comb = []
for group in combination_groups:
    combined_diffs = []
    for l in range(len(group)):
        #print(l)
        indices = range(1, len(group))
        for comb in itertools.combinations(indices, r=l):
            #print(comb)
            comb = merge_indices(group, comb)
            #print(comb)
            combined_diffs.append(np.array(comb))
    #print(group, combined_diffs)
    num_valid_comb.append(np.sum((np.all(comb < 4) for comb in combined_diffs)))
#print(num_valid_comb)
total_num_valid_comb = np.prod(num_valid_comb)
print(total_num_valid_comb)
```
