```python
input_str = """35
20
15
25
47
40
62
55
65
95
102
117
150
182
127
219
299
277
309
576"""
```

```python
with open("input_day9.txt", 'r') as f:
    input_str = f.read()
```

```python
values = input_str.splitlines()
values = [int(line) for line in values if line]
#print(values)
```

```python
import itertools

def is_sum_of_preamble(value, preamble):
    #print(value, preamble)
    pairs = itertools.combinations(preamble, r=2)
    sum_pairs = list(map(sum, pairs))
    #print(sum_pairs)
    return value in sum_pairs

def first_hit(values, preamble_len):
    for i in range(preamble_len, len(values)):
        value = values[i]
        preamble = values[(i - preamble_len):i]
        if not is_sum_of_preamble(value, preamble):
            #print(i, value)
            return value
    else:
        raise ValueError("There is no 'wrong' value in the list")
```

```python
hit = first_hit(values, preamble_len=25)
print(hit)
```

```python
def calc_encryption_weakness(values, hit):
    for start in range(len(values)):
        contiguous_set = [values[start]]
        for val in values[(start + 1):]:
            contiguous_set.append(val)
            contiguous_sum = sum(contiguous_set)
            if contiguous_sum == hit:
                smallest = min(contiguous_set)
                largest = max(contiguous_set)
                print(smallest, largest)
                encryption_weakness = smallest + largest
                return encryption_weakness
            elif contiguous_sum > hit:
                break
```

```python
encryption_weakness = calc_encryption_weakness(values, hit)
print(encryption_weakness)
```
