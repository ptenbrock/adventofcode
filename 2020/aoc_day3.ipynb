```python
trees = [[char == '#' for char in row] for row in forest.splitlines()]
trees = np.array(trees, dtype=int)
#print(trees)
```

```python
def get_num_trees(right, down):
    #print(np.array([[1, 2]]).shape)
    locations = [(loc, loc//down*right) for loc in range(0, trees.shape[0], down)]
    #print(locations)
    locations = list(map(lambda loc: (loc[0], loc[1] % trees.shape[1]), locations))
    #print(locations)
    location_trees = [trees[loc] for loc in locations]
    #print(location_trees)
    num_trees = sum(location_trees)
    #print(num_trees)
    return num_trees
```

```python
right, down = 3, 1
#print(f"Right: {right}, down: {down}")
print(get_num_trees(right, down))
```

```python
list_of_slopes = [(1, 1), (3, 1), (5, 1), (7, 1), (1, 2)]
trees_encountered = [get_num_trees(slope[0], slope[1]) for slope in list_of_slopes]
print(trees_encountered)
np.prod(trees_encountered)
```
